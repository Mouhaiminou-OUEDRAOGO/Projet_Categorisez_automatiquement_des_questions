## **Parametrage de la requête API**

```python
import requests
import pandas as pd
from datetime import datetime, timedelta

url = "https://api.stackexchange.com/2.3/questions"
tag = "python"
min_score = 50
pagesize = 50
from_date = int((datetime.now() - timedelta(days=30)).timestamp())  # Date de début (il y a 30 jours)

# Paramètres de la requête API
params = {
    "site": "stackoverflow",
    "pagesize": 50,
    "pagesize": pagesize,
    "fromdate": from_date,
    "tagged": tag,
    "min": min_score,
}

response = requests.get(url, params=params)
data = response.json()

# Récupérer les colonnes pertinentes dans les données 'items'
columns_of_interest = ['question_id', 'creation_date', 'title', 'tags', 'score', 'view_count', 'answer_count']
items = data['items']

# Créer un DataFrame en normalisant les données JSON et en spécifiant les colonnes d'intérêt
if items:
    df = pd.json_normalize(items)[columns_of_interest]
```
